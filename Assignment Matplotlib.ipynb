```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("C:/Users/Admin/Downloads/Customers - Customers.csv")
```

```python

import matplotlib.pyplot as plt

fig, axes = plt.subplots(4, 2, figsize=(15, 10))

axes[0, 0].set_title("Distribution of Age")
axes[0, 0].set_xlabel("Age")
axes[0, 0].hist(ds["Age"], color='skyblue', edgecolor='black')

axes[0, 1].set_title("Distribution of Annual Income ($)")
axes[0, 1].set_xlabel("Annual Income")
axes[0, 1].hist(ds["Annual Income ($)"], color='orange', edgecolor='black')

axes[1, 0].set_title("Distribution of Spending Score (1-100)")
axes[1, 0].set_xlabel("Spending Score")
axes[1, 0].hist(ds["Spending Score (1-100)"],  color='green', edgecolor='black')

axes[1, 1].set_title("Distribution of Family Size")
axes[1, 1].set_xlabel("Family Size")
axes[1, 1].hist(ds["Family Size"],  color='grey', edgecolor='black')

axes[2, 0].set_title("Distribution of Work Experience")
axes[2, 0].set_xlabel("Work Experience")
axes[2, 0].hist(ds["Work Experience"],  color='red', edgecolor='black')

axes[2, 1].set_title("Gender Count")
axes[2, 1].set_xlabel("Gender")
axes[2, 1].hist(ds["Gender"],  color='yellow', edgecolor='black')

plt.tight_layout()
plt.show()
```
